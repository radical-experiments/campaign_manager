```python
bookkeeping = pd.DataFrame(columns=['workflows','execution_time', 'planning_time'])


for session in work4res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [4, book_end - book_start, plan_end - plan_start]

for session in work8res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [8, book_end - book_start, plan_end - plan_start]
    
for session in work16res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [16, book_end - book_start, plan_end - plan_start]
    
for session in work32res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [32, book_end - book_start, plan_end - plan_start]
    
for session in work64res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [64, book_end - book_start, plan_end - plan_start]
    
for session in work128res4:
    bookkeeping_prof = open(session+'/bookkeper.0000.prof')
    prof_lines = bookkeeping_prof.readlines()
    for line in prof_lines:
        if 'bookkeper_start' in line:
            book_start = eval(line.split(',')[0])
        elif 'bookkeper_stopping' in line:
            book_end = eval(line.split(',')[0])
        elif 'planning_start' in line:
            plan_start = eval(line.split(',')[0])
        elif 'planning_ended' in line:
            plan_end = eval(line.split(',')[0])
    bookkeeping_prof.close()
    bookkeeping.loc[len(bookkeeping)] = [128, book_end - book_start, plan_end - plan_start]
```
